```python
import gymnasium as gym

# Create the MountainCar environment
env = gym.make("MountainCar-v0")

# Reset environment
state, info = env.reset()
print("Initial State:", state)
```

## Deep Q-Network (DQN) Implementation for CartPole

This section implements a DQN agent to solve the CartPole-v1 environment. It includes:

1.  **DQN Model**: A simple neural network using `nn.Linear` layers.
2.  **Replay Buffer**: To store and sample experiences for training.
3.  **DQNAgent**: Class to handle action selection (epsilon-greedy), experience storage, and learning.
4.  **Training Loop**: To train the agent and evaluate its performance.

```python
import torch
import torch.nn as nn
import torch.optim as optim
import random
from collections import deque

# 1. Define the DQN Model
class DQN(nn.Module):
    def __init__(self, state_size, action_size):
        super(DQN, self).__init__()
        self.fc1 = nn.Linear(state_size, 128)
        self.relu = nn.ReLU()
        self.fc2 = nn.Linear(128, 128)
        self.fc3 = nn.Linear(128, action_size)

    def forward(self, x):
        x = self.relu(self.fc1(x))
        x = self.relu(self.fc2(x))
        return self.fc3(x)
```

```python
# 2. Implement the Replay Buffer
class ReplayBuffer:
    def __init__(self, capacity):
        self.buffer = deque(maxlen=capacity)

    def push(self, state, action, reward, next_state, done):
        self.buffer.append((state, action, reward, next_state, done))

    def sample(self, batch_size):
        return random.sample(self.buffer, batch_size)

    def __len__(self):
        return len(self.buffer)
```

```python
# 3. Define the DQNAgent
class DQNAgent:
    def __init__(self, state_size, action_size, learning_rate, gamma, epsilon_start, epsilon_end, epsilon_decay, buffer_capacity, batch_size):
        self.state_size = state_size
        self.action_size = action_size
        self.gamma = gamma
        self.epsilon = epsilon_start
        self.epsilon_end = epsilon_end
        self.epsilon_decay = epsilon_decay
        self.batch_size = batch_size

        self.device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
        self.policy_net = DQN(state_size, action_size).to(self.device)
        self.target_net = DQN(state_size, action_size).to(self.device)
        self.target_net.load_state_dict(self.policy_net.state_dict())
        self.target_net.eval()

        self.optimizer = optim.Adam(self.policy_net.parameters(), lr=learning_rate)
        self.criterion = nn.MSELoss()
        self.replay_buffer = ReplayBuffer(buffer_capacity)

    def select_action(self, state):
        if random.random() < self.epsilon:
            return random.randrange(self.action_size)
        else:
            state = torch.FloatTensor(state).unsqueeze(0).to(self.device)
            with torch.no_grad():
                q_values = self.policy_net(state)
            return q_values.argmax(1).item()

    def store_experience(self, state, action, reward, next_state, done):
        self.replay_buffer.push(state, action, reward, next_state, done)

    def learn(self):
        if len(self.replay_buffer) < self.batch_size:
            return

        experiences = self.replay_buffer.sample(self.batch_size)
        states, actions, rewards, next_states, dones = zip(*experiences)

        states = torch.FloatTensor(states).to(self.device)
        actions = torch.LongTensor(actions).unsqueeze(1).to(self.device)
        rewards = torch.FloatTensor(rewards).unsqueeze(1).to(self.device)
        next_states = torch.FloatTensor(next_states).to(self.device)
        dones = torch.FloatTensor(dones).unsqueeze(1).to(self.device)

        # Compute Q-values for current states
        current_q_values = self.policy_net(states).gather(1, actions)

        # Compute max Q-value for next states from target network
        next_q_values = self.target_net(next_states).max(1)[0].unsqueeze(1)
        target_q_values = rewards + (self.gamma * next_q_values * (1 - dones))

        # Compute loss and perform backpropagation
        loss = self.criterion(current_q_values, target_q_values)
        self.optimizer.zero_grad()
        loss.backward()
        self.optimizer.step()

    def update_target_net(self):
        self.target_net.load_state_dict(self.policy_net.state_dict())

    def decay_epsilon(self):
        self.epsilon = max(self.epsilon_end, self.epsilon * self.epsilon_decay)
```

```python
# 4. Training Loop
import gymnasium as gym
import matplotlib.pyplot as plt

# Hyperparameters
STATE_SIZE = env.observation_space.shape[0]
ACTION_SIZE = env.action_space.n
LEARNING_RATE = 0.0005 # Adjusted for MountainCar
GAMMA = 0.99
EPSILON_START = 1.0
EPSILON_END = 0.01
EPSILON_DECAY = 0.999 # Adjusted for MountainCar, slower decay
BUFFER_CAPACITY = 10000
BATCH_SIZE = 64
UPDATE_TARGET_EVERY = 10
NUM_EPISODES = 2000 # Increased for MountainCar, as it's harder to solve

print(f"Environment State Size: {STATE_SIZE}, Action Size: {ACTION_SIZE}")

agent = DQNAgent(
    STATE_SIZE, ACTION_SIZE, LEARNING_RATE, GAMMA,
    EPSILON_START, EPSILON_END, EPSILON_DECAY,
    BUFFER_CAPACITY, BATCH_SIZE
)

rewards_per_episode = []

print(f"Training on device: {agent.device}")

for episode in range(1, NUM_EPISODES + 1):
    state, _ = env.reset()
    episode_reward = 0
    terminated = False
    truncated = False

    while not terminated and not truncated:
        action = agent.select_action(state)
        next_state, reward_from_env, terminated, truncated, _ = env.step(action)

        # Apply reward shaping for MountainCar
        # Add a positive reward proportional to the car's position,
        # shifted so minimum position (-1.2) gives 0 and goal (0.5) gives highest.
        reward = reward_from_env + (next_state[0] + 1.2) * 5 # Scaling factor 5

        # Store experience in replay buffer
        agent.store_experience(state, action, reward, next_state, terminated)

        # Learn from experiences
        agent.learn()

        state = next_state
        episode_reward += reward

    rewards_per_episode.append(episode_reward)

    # Update target network
    if episode % UPDATE_TARGET_EVERY == 0:
        agent.update_target_net()

    # Decay epsilon
    agent.decay_epsilon()

    if episode % 100 == 0: # Print less frequently for more episodes
        print(f"Episode {episode}/{NUM_EPISODES}, Reward: {episode_reward:.2f}, Epsilon: {agent.epsilon:.2f}")

env.close()
print("Training finished.")
```

### Training Results Visualization

Let's visualize the rewards per episode to see how the agent's performance improved over training.

```python
# Plotting the rewards
plt.figure(figsize=(10, 6))
plt.plot(rewards_per_episode)
plt.title('DQN Training Rewards per Episode (MountainCar-v0)')
plt.xlabel('Episode')
plt.ylabel('Total Reward')
plt.grid(True)
plt.show()
```
